```python
reviews_usecols_list = ['listing_id', 'date']
df_reviews20 = pd.read_csv('reviews_2020-04-20.csv', usecols = reviews_usecols_list)

listings_usecols_list = ['id', 'neighbourhood_cleansed', 'accommodates']
df_listings20 = pd.read_csv('listings_2020-04-20.csv', usecols = listings_usecols_list)
```

```python
df_reviews20['date'] = pd.to_datetime(df_reviews20['date'])                                
df_reviews20['month'] = df_reviews20['date'].dt.month
df_reviews20['year'] = df_reviews20['date'].dt.year
df_reviews20['week'] = df_reviews20['date'].dt.week

df_plot1 = df_reviews20.groupby(['year', 'week']).date.count().unstack(0)
df_plot1.head(53)
```

```python
df_plot1.loc[1:2, 2019].sum()
```

```python

fig, chart1 =  plt.subplots(2, 1)

color_dictionary = {1:'darkred', 2:'sandybrown', 3:'darkblue', 4:'green', 5:'brown', 
                    6:'mediumslateblue', 7:'black', 8:'violet', 9:'darkgoldenrod', 10:'yellow'}
#fmt_lines_dictionary ={1 : 'b-', 2 : 'y-', 3 : 'r-', 4 : 'c-', 5 : 'b-', 6 : 'y-.', 7 : 'r-.', 8 : 'c-.'}

count = 1
rate_number_reviews = []

for year in df_plot1.loc[:, 2015:].columns:
    x = df_plot1[year].index
    
    y0 = df_plot1[year] 
    chart1[0].plot(x, y0, color = color_dictionary[count], linestyle='-.', linewidth=2.5,  label = year)
    
    cumulative_sum_reviews = []
    transparency = 0.3
    for week in df_plot1[year].index:
        cumulative_sum_reviews.append(df_plot1.loc[1:week, year].sum())
        y1 = cumulative_sum_reviews
    chart1[1].plot(x, y1, color = color_dictionary[count], linestyle='dashed', linewidth=2.5,  label = year)
    transparency += 0.2
    count += 1

fig.set_size_inches(20, 10)
chart1[0].legend(shadow = True, fontsize = 15)
chart1[1].legend(shadow = True, fontsize = 15)
chart1[0].set(xlabel = 'Week', ylabel = 'Number of Reviews', 
             title = 'Number of Reviews per week',
             xlim = [1, 52], ylim=[0, 8000])

chart1[1].set(xlabel = 'Week', ylabel = 'Number of Reviews', 
             title = 'Cumulative Number of Review by the Weeks',
             xlim = [1, 52], ylim=[0, 120000])

chart1[0].grid()
chart1[1].grid()
chart1[0].set_xticks(df_plot1.index)
chart1[1].set_xticks(df_plot1.index)
```
